
# Acceleration calculation

```python
import os
import sys
```

```python
os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
from pyspark.sql import  SparkSession
```

```python
from pyspark.sql import Window
import pyspark.sql.functions as fn
```

```python
spark = SparkSession \
    .builder \
    .appName("Python Spark Telemetry data Acceleration calculation") \
    .getOrCreate()
```

```python
parquet_path="file:///home/khusheeu/Documents/KHUSHEE/TRAININGS AND CERTIFICATES/PROJECTS/Project - Rocket Telemetry data/Project-files/parquet_data/"
```

```python
df1 = spark.read.parquet(parquet_path)
```

```python
df1.show()
```

```python
df1.select('time','velocity').orderBy('time', ascending=True).show()
```

```python
df1.agg({"time": "max"}).show()
```

```python
df1.agg({"time": "min"}).show()
```

```python
val = df1.agg({"time": "max"}).collect()[0][0]
print(val)
print(val/60)
```

```python
agg_window = Window.orderBy("time")
```

```python
df1.select('time', 'velocity').withColumn('prev_velocity', fn.lag('velocity', 1).over(agg_window)).show(50)
```

```python
df1.select('time', 'velocity')\
    .withColumn('prev_time',     fn.lag('time', 1).over(agg_window))\
    .withColumn('prev_velocity', fn.lag('velocity', 1).over(agg_window))\
    .withColumn('time_diff',     fn.col('time')-fn.col('prev_time'))\
    .withColumn('velocity_diff', fn.col('velocity')-fn.col('prev_velocity'))\
    .withColumn('acceleration', fn.col('velocity_diff')/fn.col('time_diff'))\
    .show()
```

```python
df2 = df1.select('time', 'velocity')\
    .withColumn('prev_time',     fn.lag('time', 1).over(agg_window))\
    .withColumn('prev_velocity', fn.lag('velocity', 1).over(agg_window))\
    .withColumn('time_diff',     fn.col('time')-fn.col('prev_time'))\
    .withColumn('velocity_diff', fn.col('velocity')-fn.col('prev_velocity'))\
    .withColumn('acceleration', fn.col('velocity_diff')/fn.col('time_diff'))
```
